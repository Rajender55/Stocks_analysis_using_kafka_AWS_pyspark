```python
from json import loads
```

```python
consumer=KafkaConsumer('demotopic5',
    bootstrap_servers=['16.171.111.238:9092'],
                       value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
for i in consumer:
    print(i.value)
```
